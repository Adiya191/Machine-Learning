```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/Aaadityaa/Downloads/Documents/carprice.csv")
df
```

```python
dummies =pd.get_dummies(df.model)
dummies
```

```python
df_dummies=pd.concat([df,dummies],axis='columns')
df_dummies
```

```python
df_dummies.drop(['model','mercedes'],axis='columns',inplace=True)
df_dummies
```

```python
X=df_dummies.drop('sell price',axis='columns')
X
```

```python
y=df_dummies['sell price']
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
#find the price of mercedes that is 4 year old with milage 4500
model.predict([[4500,4,0,0]])
```

```python
#price of BMW that is 7 year old with milage of 86000
model.predict([[86000,7,0,1]])
```
